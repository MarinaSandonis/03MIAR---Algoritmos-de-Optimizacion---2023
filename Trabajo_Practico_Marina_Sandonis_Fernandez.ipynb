#Análisis
### ¿Qué complejidad tiene el problema?. Orden de complejidad y Contabilizar el espacio de soluciones
Teniendo en cuenta que hay 20 equipos y por tanto, se van a formar 10 partidos, tenemos 10! posibles soluciones.

Si se resolviera el problema por fuerza bruta, la complejidad del problema sería factorial. Pero al resolverse utilizando algoritmos voraces, la complejidad del mismo pasa a ser O(nlogn)

#Modelo
###1. ¿Cómo represento el espacio de soluciones?
El espacio de soluciones consta de todas las posibles asignaciones de horarios para los partidos de fútbol. Cada una de las posibles soluciones es una combinación diferente de horarios para los partidos.
Para representar este espacio de soluciones se van a utilizar DataFrames de Pandas.

###2. ¿Cuál es la función objetivo?
La función objetivo en este problema es maximizar la audiencia total. Dicha audiencia se calcula sumando la audiencia de cada partido multiplicada por el coeficiente correspondiente al horario en el que se juega dicho partido y considerando las posibles reducciones de audiencia en caso de que haya coincidencia de partidos.
$$max(\sum_{i=1}^{N}audiencia_i*PondHorario_i*CoefCoincidencia_i)$$

Donde:
* `N` Es el número de partidos
* `audiencia` Es la audiencia del partido por categoria
* `PondHorario` Es coeficiente de ponderación del horario en el que se juega el partido
* `CoefCoincidencia` Es el coeficiente de reducción de audiencia si existen coincidencias de partidos

###3. ¿Cómo implemento las restricciones?
Las restricciones del problema son las siguientes:
1. Cada partido debe programarse exactamente una vez. Para implementar esta restricción se crea un bucle 'for' para recorrer los diferentes partidos y asignar a cada uno de ellos un horario, de este modo se evita la asignación de varios horarios a un mismo partido.
2. Debe haber obligatoriamente un partido el viernes y uno el lunes. Para implementar esta restricción se asignan estos horarios a los dos partidos de menor audiencia. Así se garantiza que estos horarios, que son los que tienen un coeficiente de mayor reducción de audiencia, no quedan vacíos y tampoco se pierde mucha audiencia.


#Diseño
- ¿Que técnica utilizo? ¿Por qué?

Se va a utilizar una técnica de algoritmos voraces para resolver el problema. Por tanto, se va a dividir el problema en diferentes etapas y en cada una de ellas se va a seleccionar la solución más adecuada, es decir, la solución que maximice la audiencia. En este caso concreto, cada una de las etapas se corresponde con un partido, para el que se va a intentar maximizar la audiencia. De este modo se asegura que no se va a programar cada partido más de una vez y se trata de alcanzar el mejor resultado de audiencia posible mediante la maximización de la audiencia de cada partido.

Para implementar la técnica de algoritmos voraces lo que se ha hecho es ordenar el dataframe de la audiencia de menor a mayor. Los partidos de menor audiencia se van a asignar a los horarios del lunes y viernes y así se cumple con la restricción de que exista un partido en esos horarios y se logra perder la menor audiencia posible, pues se trata de los horarios con mayor reducción de la audiencia.

En caso de que se la audiencia máxima de varios partidos se consiga para el mismo horario, la audiencia de cada uno de los partidos coincidentes se verá reducida. Por ejemplo, en caso de que coincidan dos partidos, la audiencia de cada uno de ellos se reduce un 25%.

A continuación, se muestra el algoritmo propuesto y la solución obtenida para la jornada proporcionada en el enunciado del trabajo.

```
#Definición de los datos

import pandas as pd

pond_cat_partidos = pd.Series({'AA': 2, 'AB':1.3, 'AC':1, 'BB':0.9, 'BC': 0.75, 'CC':0.47, 'BA':1.3, 'CA':1, 'CB':0.75})
display(pond_cat_partidos)

pond_horarios = pd.Series({'V20h':0.4, 'S12h':0.55, 'S16h':0.7, 'S18h':0.8, 'S20h': 1, 'D12h':0.45, 'D16h':0.75, 'D18h':0.85, 'D20h':1, 'L20h':0.4})

display(pond_horarios)

coef_coincidencia =  pd.DataFrame([0, 0.25, 0.45, 0.6, 0.7, 0.75, 0.78, 0.8, 0.8], columns=['%'])
display(coef_coincidencia)

nombres = ['Real Madrid','R.Sociedad','Barcelona','Celta','Valencia','Athletic','Villarreal','Alavés','Levante',
           'Espanyol','Sevilla','Betis','Atlético','Getafe', 'Mallorca','Eibar','Leganés','Osasuna','Granada','Valladolid' ]
catego = ['A', 'A', 'A','B','B','B','B','B','B','B','B','B','B','B','C','C','C','C','C','C']
equipos = pd.DataFrame(catego, index = nombres, columns=['Categoría'])
display(equipos)
```

```
#Partidos de la jornada.
partidos_jornada = [['Celta', 'Real Madrid'], ['Valencia', 'R.Sociedad'], ['Mallorca', 'Eibar'],
                    ['Athletic', 'Barcelona'], ['Leganés', 'Osasuna'], ['Villarreal', 'Granada'],
                    ['Alavés', 'Levante'], ['Espanyol', 'Sevilla'], ['Betis', 'Valladolid'], ['Atlético', 'Getafe']]
```

```
#Ordenar partidos por audiencias
import numpy as np
def ordenar_partidos_audiencia(equipos_partidos, equipos, pond_cat_partidos):
  audiencia = {}
  for i in range(len(equipos_partidos)):
    aux = equipos_partidos[i]
    catego = str(equipos.loc[aux[0]][0])+str(equipos.loc[aux[1]][0])
    name_partido = str(aux[0])+'-'+str(aux[1])
    audiencia[name_partido] = pond_cat_partidos.loc[catego]
  audiencia = pd.Series(audiencia).sort_values(ascending=True)
  return audiencia


partidos_ordenados = ordenar_partidos_audiencia(partidos_jornada, equipos, pond_cat_partidos)
print(partidos_ordenados)
```

```
horarios_asignados = pd.Series({'V20h':0, 'S12h':0, 'S16h':0, 'S18h':0, 'S20h': 0, 'D12h':0, 'D16h':0, 'D18h':0, 'D20h':0, 'L20h':0})
audiencias_finales = pd.DataFrame(columns=['Horario', 'Audiencia'])
audiencia_tot=0
cont = 0
mejor_audiencia = 0
mejor_lunes = 0
mejor_viernes = 0

for partido, audiencia in zip(partidos_ordenados.index, partidos_ordenados.values):
  mejor_audiencia = 0
  num_coinci = 0
  #Para cada partido se selecciona el mejor horario (el que más audiencia tiene)
  #Aseguramos que al horario del lunes y del viernes se les asigna un partido
  if horarios_asignados.loc['V20h'] ==0:
    mejor_horario = 'V20h'
    num_coinci = horarios_asignados['V20h']
    mejor_audiencia = pond_horarios['V20h']*audiencia*(1- coef_coincidencia.iloc[num_coinci, 0])
  elif horarios_asignados.loc['L20h'] ==0:
    mejor_horario = 'L20h'
    num_coinci = horarios_asignados['L20h']
    mejor_audiencia = pond_horarios['L20h']*audiencia*(1- coef_coincidencia.iloc[num_coinci, 0])
  else: #Para el resto de partidos seleccionamos el horario que maximice la audiencia
    for horario in pond_horarios.index:
      audiencia_actual = pond_horarios[horario]*audiencia*(1- coef_coincidencia.iloc[horarios_asignados[horario], 0])

      if audiencia_actual.astype(float) > mejor_audiencia:
        mejor_audiencia = audiencia_actual
        mejor_horario = horario
        num_coinci = horarios_asignados[horario]

  #Una vez se tiene el mejor horario para el partido se asigna
  if num_coinci > 0: #Si ha habido una coincidencia se resta la audiencia correspondiente al partido previamente asignado
    repe = audiencias_finales[audiencias_finales["Horario"] == mejor_horario]
    audiencias_finales.loc[repe.index, 'Audiencia'] = audiencias_finales.loc[repe.index, 'Audiencia']*(1- coef_coincidencia.iloc[num_coinci, 0])
    audiencia_tot =audiencia_tot - repe.loc[repe.index,'Audiencia'].iloc[0] + audiencias_finales.loc[repe.index, 'Audiencia'].iloc[0]
  audiencia_tot = audiencia_tot + mejor_audiencia
  horarios_asignados.loc[mejor_horario] = horarios_asignados.loc[mejor_horario] +1
  audiencias_finales.loc[partido] =[mejor_horario, mejor_audiencia]

print(audiencia_tot)
display(audiencias_finales)
```
